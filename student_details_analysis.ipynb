```python
import numpy as np
import pandas as pd
import seaborn as sns
import  matplotlib.pyplot as plt
```

```python
#Data_laoding
df=pd.read_csv("Student_details.csv")
df
```

```python
#Total marks
df["Total_marks"]=df[["Maths","Physics","Chemistry","Computer"]].sum(axis=1)
df.head()
```

```python
#percentage
df["Percentage"]=df["Total_marks"]/4
df
```

```python

# Grade Assignment
def assign_grade(p):
    if p >= 85:
        return "A"
    elif p >= 70:
        return "B"
    elif p >= 55:
        return "C"
    else:
        return "D"

df["Grade"] = df["Percentage"].apply(assign_grade)
df
```

```python
# -----------------------------
# Statistical Analysis
# -----------------------------
```

```python
print("Average Percentage:",np.mean(df['Percentage']))
print("Maximum Percentage:",np.max(df['Percentage']))
print("Minimum Percentage:",np.min(df['Percentage']))
```

```python
#--------------------------------------------Visualization-------------------------------------------------
```

```python
# Study Hours vs Percentage
plt.figure(figsize=(10,15))
sns.relplot(data=df,x="Study_Hours",y="Percentage",hue="Grade",kind="scatter")
plt.xlabel('Student Hours')
plt.ylabel('Percentage')
plt.title('Student Hours vs Percentage')
plt.show()
```

```python
# Attendance vs Percentage
plt.figure(figsize=(10,15))
sns.relplot(data=df,x="Attendance",y="Percentage",hue="Grade",kind="scatter")
plt.xlabel('Attendance')
plt.ylabel('Percentage')
plt.title('Attendance vs Percentage')
plt.show()
```

```python
# Gender Comparison
plt.figure()
sns.boxplot(x="Gender", y="Percentage", data=df)
sns.relplot(x="Gender", y="Percentage", data=df,hue="Grade",kind="scatter")
plt.title("Gender vs Performance")
plt.show()
```

```python
# Subject Average Comparison
sub_avg=df[['Maths','Physics','Chemistry','Computer']].mean()
plt.figure()
sub_avg.plot(kind='bar')
```

```python
# Correlation Heatmap
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
